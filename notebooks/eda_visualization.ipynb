```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Thiết lập style
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))
growth = pd.read_csv("q1_growth.csv")
# Chuyển Series thành DataFrame và sắp xếp giảm dần để biểu đồ đẹp hơn
growth_df = growth.reset_index().sort_values(by='price_change_pct', ascending=False)

# Vẽ Bar Chart
sns.barplot(data=growth_df, x='symbol', y='price_change_pct', palette='viridis')

# Trang trí
plt.title('Tăng trưởng giá cổ phiếu (01/06 - 07/10/2025)', fontsize=14, fontweight='bold')
plt.xlabel('Mã cổ phiếu', fontsize=12)
plt.ylabel('Tổng tăng trưởng (%)', fontsize=12)
plt.xticks(rotation=45)

# Thêm số liệu trên đầu mỗi cột
for index, row in enumerate(growth_df['price_change_pct']):
    plt.text(index, row + 0.5, f'{row:.1f}', color='black', ha="center", fontsize=9)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 12))
sns.set_theme(style="whitegrid")
result = pd.read_csv("q2_result.csv")
# Biểu đồ 1: Khối lượng giao dịch (Avg_Volume)
sns.barplot(data=result.reset_index(), x='symbol', y='Avg_Volume', hue='Period', 
            ax=axes[0], palette='Blues')
axes[0].set_title('So sánh Khối lượng Giao dịch: Trước vs Sau 07/10/2025', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Khối lượng trung bình')
axes[0].tick_params(axis='x', rotation=45)

# Biểu đồ 2: Vận tốc tăng trưởng (Avg_Daily_Return)
sns.barplot(data=result.reset_index(), x='symbol', y='Avg_Daily_Return', hue='Period', 
            ax=axes[1], palette='coolwarm')
axes[1].set_title('So sánh Vận tốc Tăng trưởng: Trước vs Sau 07/10/2025', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Lợi nhuận trung bình (%)')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.set_theme(style="whitegrid")
momentum_df = pd.read_csv("q3_momentum.csv")
# Tính lợi nhuận tích lũy bằng hàm cumsum()


sns.lineplot(data=momentum_df, x="date", y="Bank_Momentum", label="Bank")
sns.lineplot(data=momentum_df, x="date", y="VNIndex_Momentum", label="VNINDEX")

plt.title("Bank Momentum vs VNINDEX Momentum")
plt.xlabel("Date")
plt.ylabel("Daily Return")

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.set_theme(style="whitegrid")

# Xóa các dòng có NaN nếu có (tránh lỗi vẽ)
risk_metrics = pd.read_csv("q4_risk.csv")

# Xóa các dòng có NaN nếu có (tránh lỗi vẽ)
risk_df = risk_metrics.dropna().reset_index()



# Vẽ các dấu chấm (Scatter)
sns.scatterplot(data=risk_df, x='Volatility', y='Max_Drawdown', hue='symbol', s=150, palette='tab20')

# Gắn nhãn tên từng mã cổ phiếu (Dùng .iloc để lấy giá trị chính xác)
for i in range(len(risk_df)):
    plt.annotate(
        risk_df['symbol'].iloc[i], # Lấy tên mã cổ phiếu
        (risk_df['Volatility'].iloc[i], risk_df['Max_Drawdown'].iloc[i]), # Tọa độ x, y của chấm
        xytext=(8, 0), # Đẩy chữ nhích sang phải 8 pixel để khỏi đè lên cái chấm
        textcoords='offset points', 
        fontsize=11, 
        fontweight='bold', # Bôi đậm tên cho dễ nhìn
        color='darkblue'   # Màu chữ
    )

plt.title('Ma Trận Rủi Ro: Độ biến động (Volatility) vs Sụt giảm tối đa (Max Drawdown)', fontsize=14, fontweight='bold')
plt.xlabel('Volatility (Độ biến động hàng năm %)', fontsize=12)
plt.ylabel('Max Drawdown (Sụt giảm tối đa %)', fontsize=12)

# Ẩn cái hộp chú thích (Legend) đi cho đỡ chật diện tích vì đã có tên trên chấm rồi
plt.legend([],[], frameon=False) 

plt.tight_layout()
plt.show()
```
