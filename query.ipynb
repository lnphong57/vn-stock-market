```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_json("D:/Projects/DS/data/processed/price_history_processed.json")
df['Ngày'] = pd.to_datetime(df['Ngày'], format='%d/%m/%Y')
stockCode = [
    "ACB", "BID", "CTG", "EIB","EVF", "HDB", "KLB", "LPB", "MBB", "MSB", "NAB", 
    "OCB", "PVF", "SHB", "SSB", "STB", "TCB", "TPB", "VAB", "VCB", "VIB", "VPB"
]
```

Chọn lịch sử giá của mã CP

```python

while True:
    stock_id = str(input("Nhập mã CP bạn muốn xem: ")).upper()
    if stock_id in stockCode:
        #bỏ côt symbol
        try:
            df.drop(columns=["symbol"], inplace=True)
        except KeyError:
            pass

        #gắn title
        print(f"-----{stock_id.upper()} - Lịch sử giá-----")
        print(df)
        break
    else:
        print(" Mã không hợp lệ, vui lòng chọn lại")
```
